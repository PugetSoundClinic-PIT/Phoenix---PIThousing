```python
# Set the path to the ChromeDriver executable as an environment variable
os.environ["webdriver.chrome.driver"] = r'C:\Users\jaini\College Stuff- UW\OneDrive - UW\UW info Research\chromedriver.exe'

# Now, you can directly create a Chrome webdriver instance without specifying the executable path
driver = webdriver.Chrome()

# Set the path to the ChromeDriver executable as an environment variable
os.environ["webdriver.chrome.driver"] = r'C:\Users\jaini\College Stuff- UW\OneDrive - UW\UW info Research\chromedriver.exe'

# Now, you can directly create a Chrome webdriver instance without specifying the executable path
driver = webdriver.Chrome()

# setting path to download directory

download_directory = r"C:\Users\jaini\College Stuff- UW\OneDrive - UW\UW info Research\downladed_files"

from selenium import webdriver
import os
import time

os.environ["webdriver.chrome.driver"] = r'C:\Users\jaini\College Stuff- UW\OneDrive - UW\UW info Research\chromedriver.exe'

# Now, you can directly create a Chrome webdriver instance without specifying the executable path
chrome_driver = webdriver.Chrome()

# Open the website
chrome_driver.get('https://recorder.maricopa.gov/recdocdata/GetRecDataRecentPgDn.aspx?rec=0&suf=&nm=&bdt=1%2f1%2f2018&edt=3%2f6%2f2024&cde=DE&max=20&res=True&doc1=DE&doc2=&doc3=&doc4=&doc5=')

download_directory = r"C:\Users\jaini\College Stuff- UW\OneDrive - UW\UW info Research\downladed_files"

# URLs to visit
base_url = 'https://recorder.maricopa.gov/recdocdata/'
recording_numbers = [
    '20180001030', '20180002080', '20180002082', '20180002100', '20180002919',
    '20180002972', '20180003451', '20180004495', '20180005770', '20180007458',
    '20180007727', '20180007771', '20180009383', '20180009807', '20180010268',
    '20180010694', '20180011034', '20180011164', '20180011966', '20180015239'
]

# Function to download PDFs for each recording number
def download_pdfs(recording_numbers):
    for number in recording_numbers:
        try:
            # Construct the URL
            url = f"{base_url}UnofficialPdfDocs.aspx?rec={number}&pg=1&cls=RecorderDocuments&suf="
            
            # Visit the URL
            driver.get(url)
            time.sleep(3)  # Add a delay to allow the page to load
            
            # Find the link to download the PDF and click it
            download_link = driver.find_element_by_xpath('//a[text()="Download"]')
            download_link.click()
            
            # Add a delay to ensure the download completes
            time.sleep(5)
            
            print(f"PDF for recording number {number} downloaded successfully.")
        
        except Exception as e:
            print(f"Error downloading PDF for recording number {number}: {e}")

# Download PDFs for each recording number
download_pdfs(recording_numbers)

# Quit the WebDriver
driver.quit()
```
